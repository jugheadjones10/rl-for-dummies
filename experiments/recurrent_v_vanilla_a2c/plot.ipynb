```python
import os

import matplotlib.pyplot as plt
import pandas as pd
```

#### MiniGrid-Empty-6x6-v0


```python
# Create a figure for individual seed plots
plt.figure(figsize=(15, 6))

vanilla_all_dfs = []
# For each csv result file in results directory, plot the results
for dir in os.listdir("results"):
    if "vanilla" in dir and "Empty-6x6" in dir:
        seed = dir.split("_")[-1]
        for file in os.listdir(os.path.join("results", dir)):
            if file.endswith(".csv"):
                df = pd.read_csv(os.path.join("results", dir, file))
                plt.plot(df["step"], df["avg_score"], label=f"seed {seed}")
                vanilla_all_dfs.append(df)

plt.title("Individual Seed Results for Vanilla A2C on MiniGrid-Empty-6x6-v0")
plt.xlabel("Training Steps")
plt.ylabel("Average Score")
plt.legend()
plt.show()
```

```python
# Create a figure for individual seed plots
plt.figure(figsize=(15, 6))

recurrent_all_dfs = []
# For each csv result file in results directory, plot the results
for dir in os.listdir("results"):
    if "recurrent" in dir and "Empty-6x6" in dir:
        seed = dir.split("_")[-1]
        for file in os.listdir(os.path.join("results", dir)):
            if file.endswith(".csv"):
                df = pd.read_csv(os.path.join("results", dir, file))
                plt.plot(df["step"], df["avg_score"], label=f"seed {seed}")
                recurrent_all_dfs.append(df)

plt.title("Individual Seed Results for Recurrent A2C on MiniGrid-Empty-6x6-v0")
plt.xlabel("Training Steps")
plt.ylabel("Average Score")
plt.legend()
plt.show()
```

```python
# Create a new figure for the average plot
plt.figure(figsize=(15, 6))


def plot_average(all_dfs, color, label):
    # Group by step and calculate mean
    combined_df = pd.concat(all_dfs)
    avg_df = combined_df.groupby("step").mean().reset_index()

    # Now step is a column again after reset_index()
    plt.plot(avg_df["step"], avg_df["avg_score"], color + "-", linewidth=3, label=label)

    # Optional: Add standard deviation
    std_df = combined_df.groupby("step").std().reset_index()
    plt.fill_between(
        avg_df["step"],
        avg_df["avg_score"] - std_df["avg_score"],
        avg_df["avg_score"] + std_df["avg_score"],
        color=color,
        alpha=0.2,
        label="±1 std dev",
    )


plot_average(vanilla_all_dfs, "r", "Vanilla A2C")
plot_average(recurrent_all_dfs, "b", "Recurrent A2C")

plt.title("Average Results Across Seeds for MiniGrid-Empty-6x6-v0")
plt.xlabel("Training Steps")
plt.ylabel("Average Score")
plt.legend()
plt.show()
```
